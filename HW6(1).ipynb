```python
%matplotlib inline
import scipy.constants
import numpy as np
import matplotlib.pyplot as plt

def n_mgf2(lambda0):
    '''
    calculate index of refraction of Magnesium Fluoride (MgF2)
    Args:
        lambda0: wavelength in meters
    Returns:
        index of refraction
    '''
    x = lambda0*1e6 
    n=np.sqrt(1+0.27620+0.60967/(1-(0.08636/x)**2)+0.0080/(1-(18.0/x)**2)+2.14973/(1-(25.0/x)**2))
    return n


def sag(radial_position, radius_of_curvature):
    '''
    calculate the sag for a spherical lens
    Arg:
        radial_position: transverse distance from center of lens [m]
        radius_of_curvature: spherical radius for the lens surface [m]
    Returns
        distance between lens surface and an optical flat.
    '''
    s = radius_of_curvature - np.sqrt(radius_of_curvature**2-radial_position**2) 
    return s
```
