```python
%matplotlib inline
import scipy.constants
import scipy.special
import numpy as np
import matplotlib.pyplot as plt


def single_slit_diffraction(lambda0, a, theta):
    '''
    Fraunhofer diffration from single slit at an angle
    Args:
        lambda0: wavelength of light (m)
        a: width of slit             (m)
        theta: angle from normal     (radians)
    Returns
        Normalized irradiance at angle theta
    '''
    k = 2 * np.pi /lambda0
    beta = k * a * np.sin(theta)/2
    return np.sinc(a*np.sin(theta)/lambda0)**2


def single_slit_diffraction_screen(lambda0, a, D, y):
    '''
    Fraunhofer diffration from single slit on a screen
    Args:
        lambda0: wavelength of light (m)
        a: width of slit             (m)
        D: slit to screen distance   (m)
        y: transverse position on screen (m)
    Returns
        Normalized irradiance at position y
    '''
    return np.sinc(a*y/lambda0/D)**2


def Airy(x):
    '''
    Compute J_1(x)/x
    '''
    if abs(x) < 1e-8:
        return 1
    return (2*scipy.special.j1(x)/x)**2
Airy = np.vectorize(Airy)


def circular_hole_diffraction(lambda0, a, theta):
    '''
    Fraunhofer diffration from hole at an angle
    Args:
        lambda0: wavelength of light (m)
        a: radius of aperture        (m)
        theta: angle from normal     (radians)
    Returns
        Normalized irradiance at angle theta
    '''
    return Airy(2 * np.pi /lambda0 * a * theta)


def double_slit_diffraction(lambda0, b, h, theta):
    '''
    Fraunhofer diffration from double slit at an angle
    (Fowles 5.27)
    Args:
        lambda0: wavelength of light (m)
        b: width of slits             (m)
        h: center to center distance  (m)
        theta: angle from normal     (radians)
    Returns
        Normalized irradiance at angle theta
    '''
    k = 2 * np.pi /lambda0
    beta = k * b * np.sin(theta)/2
    gamma = k * h * np.sin(theta)/2
    return np.sinc(beta/np.pi)**2 * np.cos(gamma)**2


def rectangular_diffraction(lambda0, a, b, thetax, thetay):
    '''
    Fraunhofer diffration from a rectangle at an angle
    Args:
        lambda0: wavelength of light         (m)
        a: horizontal width of rectangle     (m)
        b: vertical height of rectangle      (m)
        thetax: horizontal angle from normal (radians)
        thetay: vertical angle from normal   (radians)

    Returns
        Normalized irradiance at thetax, thetay
    '''
    ix = single_slit_diffraction(lambda0,a,thetax)
    iy = single_slit_diffraction(lambda0,b,thetay)
    return ix*iy
```
